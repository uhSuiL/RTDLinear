```python
import pandas as pd
import seaborn as sns

sns.set()

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/processed/data_filter.csv', index_col=0)
```

```python
data
```

```python
df = (data - data.mean()) / data.std()
df
```

```python
a1 = ((df > 1) & (df <=2)).sum() / df.shape[0]
a2 = ((df > 2) & (df <=3)).sum() / df.shape[0]
a3 = (df > 3).sum() / df.shape[0]

# 定义百分数格式化函数
def percent_formatter(x, pos):
    return f'{x * 100:.0f}%'  # 将小数转换为百分数格式

# 设置 Y 轴的格式化器
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(20, 10))
plt.title('Anomaly Proportions over 149 ATMs', fontsize=30, fontfamily='Times New Roman', fontweight="bold")
plt.xlabel('ATMs', fontsize=20, fontfamily='Times New Roman')
# plt.ylabel('Anamoly Proportion', fontsize=21, fontfamily='Times New Roman')
plt.yticks(fontsize=20, fontfamily='Times New Roman')
plt.xticks(fontfamily='Times New Roman')
plt.gca().yaxis.set_major_formatter(FuncFormatter(percent_formatter))
plt.bar(range(df.shape[1]), a1, label='θ < μ ≤ 2θ', color = 'skyblue')
plt.bar(range(df.shape[1]), a2, bottom=a1, label='2θ < μ ≤ 3θ')
plt.bar(range(df.shape[1]), a3, bottom=a1 + a2, label='3θ < μ')

plt.ylim(0.05, 0.22)
plt.legend(fontsize=20)
```
