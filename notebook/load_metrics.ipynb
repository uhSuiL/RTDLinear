```python
import pandas as pd
import numpy as np
```

```python
types = ['mean', 'median']

df_list = []
for c in range(4):
    df1 = pd.read_csv(f'DLmodel_metrics({types[0]})_cluster_{c}.csv', index_col=0)
    svr1 = pd.read_csv(f'SVR_metrics({types[0]})_cluster_{c}.csv', index_col=0).rename(columns={f'cluster {c}': 'SVR'})
    mlp1 = pd.read_csv(f'MLP_metrics({types[0]})_cluster_{c}.csv', index_col=0).rename(columns={f'cluster {c}': 'MLP'})
    arima1 = pd.read_csv(f'./ARIMA/cluster_{c}_metrics.csv').iloc[1:].mean().rename('ARIMA')
    df1 = pd.concat([df1, svr1.T, pd.DataFrame(arima1.drop('mse')).T, mlp1.T]).drop('dstat', axis=1)
    df_list.append(df1)
	
metrics_table = pd.concat(df_list, keys=[f'Cluster {c+1}' for c in range(4)])
	
metrics_table
```

```python
benchmark_models = ['RTDLinear', 'DLinear', 'GRU', 'LSTM', 'SVR', 'ARIMA', 'MLP']
ablation_models = ['RTDLinear', 'TDLinear', 'DLinear']
```

```python
metrics_table.columns = [col.upper() for col in  metrics_table.columns]
```
